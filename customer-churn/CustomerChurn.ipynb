```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Customer Churn.csv")
df.head(21)

```

```python
def conv(value):
    if value== 1:
        return "yes"
    else:
        return "no"
         
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)  
df.head(25)
```

```python
plt.figure(figsize=(3,4))
ax= sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0])

plt.title("Count of Customers by Churn", fontsize = "10")
plt.show()
```

```python
gb= df.groupby('Churn').aggregate({'Churn':'count'})
gb
```

```python
plt.figure(figsize= (3,4))
plt.pie(gb['Churn'], labels = gb.index, autopct= "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize= "10")
plt.show()
```

```python
plt.figure(figsize= (3,4))
ax= sns.countplot(x= "gender", data = df, hue= "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Gender", fontsize= "10")
plt.show()
```

```python
plt.figure(figsize= (3,4))
ax= sns.countplot(x= "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizens", fontsize= "10")
plt.show()
```

```python
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate percentages
percent_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

# Step 2: Plot stacked bar chart
fig, ax = plt.subplots(figsize=(4, 5))

bottom_vals = [0] * len(percent_data)
categories = percent_data.columns

for category in categories:
    ax.bar(
        percent_data.index,
        percent_data[category],
        bottom=bottom_vals,
        label=category
    )
    # Add percentage labels
    for i, val in enumerate(percent_data[category]):
        if val > 0:
            ax.text(
                x=percent_data.index[i],
                y=bottom_vals[i] + val / 2,
                s=f"{val:.1f}%",
                ha='center',
                va='center',
                fontsize=8
            )
    bottom_vals = [i + j for i, j in zip(bottom_vals, percent_data[category])]

# Step 3: Customize plot
ax.set_xticks([0, 1])
ax.set_xticklabels(['Non-Senior', 'Senior'])
ax.set_ylabel("Percentage")
ax.set_title("Churn Distribution by SeniorCitizen", fontsize=10)
ax.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
# List of categorical columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Subplot layout: 3 columns per row
n_cols = 3
n_rows = -(-len(cols) // n_cols)  # Ceiling division

# Set figure size dynamically
plt.figure(figsize=(5 * n_cols, 4 * n_rows))

for idx, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, idx)
    ax = sns.countplot(data=df, x=col, hue="Churn")
    plt.title(f"{col} vs Churn", fontsize=10)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=20)
    # Add count labels
    for container in ax.containers:
        ax.bar_label(container, fontsize=8)

plt.tight_layout()
plt.show()
```
